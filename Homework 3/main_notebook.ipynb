```python
def count_unigram(f, p = True):
    f = ' '.join(f.split())
    f = f.lower()
    unigram_data = {}
    for i in f:
        if p and not bool(re.search('[а-я ]', i)):
            continue
        if i not in unigram_data:
                unigram_data[i] = 0
        unigram_data[i] += 1
    return sorted(unigram_data.items(), reverse=True, key=lambda x:x[1])

unigram_corp = count_unigram(big_text)
unigram_dict = { i : j for i, j in unigram_corp}
```

```python
def count_bigram(f):
    f = ' '.join(f.split())
    f = f.lower()
    bigram_data = {}
    for i in range(len(f) - 1):
        if not bool(re.search('[а-я ]', f[i])) or not bool(re.search('[а-я ]', f[i + 1])):
            continue
        s = f[i] + f[i + 1]
        if s not in bigram_data:
                bigram_data[s] = 0
        bigram_data[s] += 1
    return sorted(bigram_data.items(), reverse=True, key=lambda x:x[1])


bigram_corp = count_bigram(big_text)
bigram_dict = { i : j for i, j in bigram_corp}
bigram_test = count_bigram(test)
```

```python
def get_text(text, tr_bigram):
    res = ""
    for i in text:
        if i in tr_bigram:
            res = res + tr_bigram[i]
        else:
            res = res + i
    return res
```

```python
sum = 0
for i, j in bigram_corp:
    sum += j
sum
```

```python
def get_met(text):
    global bigram_dict   
    res = 0
    for i in range(len(text) - 1):
        s = text[i] + text[i + 1]
        if text[i] + text[i + 1] in bigram_dict:
            res += log(bigram_dict[s]) - log(sum)
        else:
            res -= log(len(big_text))
    return res
```

```python
def mcmc(test, unigram_corp, unigram_test):

    alph = [i for i, _ in unigram_corp]
    alph_test = [i for i, _ in unigram_test]
    d = { i : j for i, j in zip(alph_test, alph)}

    txt = get_text(test,  d)
    cur_met = get_met(txt)
    for i in range(150000):
        x = random.choices(range(len(alph_test)))[0]
        y = random.choices(range(len(alph)))[0]
        while y == x:
            y = random.choices(range(len(alph)))[0]

        alph[x], alph[y] =  alph[y], alph[x]
        d = { i : j for i, j in zip(alph_test, alph)}
        new_txt = get_text(test, d)
        assert len(new_txt) == len(test)
        new_met = get_met(new_txt)
        tr = min(1, exp(new_met - cur_met))
        if random.random() < tr:
            txt = new_txt
            cur_met = new_met
            continue
        alph[x], alph[y] =  alph[y], alph[x]
    return txt
```

```python
print("Результат:")
print(mcmc(test, unigram_corp, unigram_test))
```

Получились очень хорошие результаты! Текст получилось расшифровать полностью
